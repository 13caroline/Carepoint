```python
#2008-11-11 13:23:44
from random import randint

def generate_date():
    string = ""
    
    # Data de criacao
    ano = randint(2022,2023)
    mes = randint(1,12)
   
    if mes == 1 or mes == 3 or mes == 5 or mes == 7 or mes == 8 or mes == 10 or mes == 12:
        dia = randint(1,31)
    elif mes == 2:
        #fuck you 29
        dia = randint(1,28)
    else:
        dia = randint(1,30)

    hora = randint(0,23)
    minuto = randint(0,59)
    segundo = randint(0,59)
    
    string+=str(ano)+"-"
    if mes < 10: 
        string+= "0" + str(mes)+"-"
    else:
        string+= str(mes)+"-"
    
    if dia < 10: 
        string+= "0" + str(dia)+" "
    else:
        string+= str(dia)+" "
    
    if hora < 10: 
        string+= "0" + str(hora)+":"
    else:
        string+= str(hora)+":"
    
    if minuto < 10: 
        string+= "0" + str(minuto)+":"
    else:
        string+= str(minuto)+":"
    
    if segundo < 10: 
        string+= "0" + str(segundo)
    else:
        string+= str(segundo)
    
    return string
```

```python
from random import randint

def generate_aux(length):

    chars = "abcdefghijlmnopqrstuvxz "
    from os import urandom

    return "".join(chars[c % len(chars)] for c in urandom(length))

def generate_content():

    num = randint(1,160)

    return generate_aux(num)
```

```python
def get_sps():
    file_user = open("user.sql", "r")

    sp = []
    i = 0
    for row in file_user:
   
        if i == 0:
            i+=1
        else:
            x = row.split(",")
            if x[2] == "password" and x[16] == '3':
                y = x[10].split('(')
                sp.append(y[1])

    for i in sp:
        print(i)
        
    file_user.close()
    return sp
```

```python
def get_consumers():
    file_user = open("user.sql", "r")

    consumer = []
    i = 0
    for row in file_user:
   
        if i == 0:
            i+=1
        else:
            x = row.split(",")
            if x[2] == "password" and x[16] == '2':
                y = x[10].split('(')
                consumer.append(y[1])

    for i in consumer:
        print(i)
        
    file_user.close()
    return consumer
```

```python
from random import randint

def create_messages(i,file,sps,consumers):
    j = 1
    tamanho_sps = len(sps)
    tamanho_consumers = len(consumers)
    
    while j < i+1:
        string = "INSERT INTO Message(idMessage,date,content,idGive,idReceive)VALUES(" + str(j) + ",'"
        string += generate_date() + "','"
        string += generate_content() + "',"
        
        choice = randint(0,1)
        if choice == 0:
            num = randint(0,tamanho_consumers - 1)
            string += str(consumers[num]) + ","
            num2 = randint(0,tamanho_sps - 1)
            string += str(sps[num2]) + ");\n"
        else:
            num2 = randint(0,tamanho_sps - 1)
            string += str(sps[num2]) + ","
            num = randint(0,tamanho_consumers - 1)
            string += str(consumers[num]) + ");\n"
        
        file.write(string)
        j+=1
    
    file.close()
```

```python
file = open("message.sql", "a")
file.write("USE PI;\n");

#1 admin | 2 consumer | 3 sp | 4 company 
sps = get_sps()
consumers = get_consumers()

create_messages(600,file,sps,consumers)

file.close()
```
