```python
import names

def generate_name(sex):
    if sex == 'M':
        return names.get_full_name(gender='male')
    elif sex == 'F':
        return names.get_full_name(gender='female')
    else:
        return names.get_full_name()
```

```python
def generate_password(length):

    chars = "ABCDEFGHJKLMNPQRSTUVWXYZ0123456789"
    from os import urandom

    return "".join(chars[c % len(chars)] for c in urandom(length))
```

```python
import random
import string
from random import randint

def random_char(y):
       return ''.join(random.choice(string.ascii_lowercase ) for x in range(y))

def generate_email():
    num = randint(1,3)

    if num == 1:
        num2 = randint(5,13)
        return random_char(num2)+"@gmail.com"
    elif num == 2:
        num2 = randint(5,13)
        return random_char(num2)+"@outlook.com"
    else:
        num2 = randint(5,13)
        return random_char(num2)+"@hotmail.com"
```

```python
from random import randint

def create_sp(i,file):
    string_user = "INSERT INTO User(idUser,name,password,email,phoneNumber,sex,type,createdAt,lastActivity,active,idLocation)VALUES("
    string_user += str(i)+","
    
    sex = generate_sex()
    
    string_user += "'" + generate_name(sex) +"',"
    num_pw = randint(5,20)
    string_user += "'" + generate_password(num_pw) +"',"
    string_user += "'" + generate_email() +"',"
    
    phoneNumber = generate_phoneNumber()
    if phoneNumber == "NULL":
        string_user +=  phoneNumber +","
    else:
        string_user += "'" + phoneNumber+"',"
        
    string_user += "'" + sex+"',"
    string_user += str(3) +","
    string_date = generate_date().split("|")
    
    string_user += "'" + string_date[0]+"',"
    string_user += "'" + string_date[1]+"',"
    string_user +=  str(generate_active())+","
    string_user +=  str(generate_location())+");\n"
    
    file.write(string_user)

    string_sp = "INSERT INTO ServiceProvider(idSP,description,dateOfBirth,endSub,distance,qualifications,idSubscription)VALUES("
    string_sp += str(i)+","
    string_sp += "'" + generate_description() +"',"
    
    dateOfBirth = generate_dateOfBirth()
    if dateOfBirth == "NULL":
        string_sp +=  dateOfBirth +","
    else:
        string_sp += "'" + dateOfBirth+"',"
    
    endSub = generate_endSub()
    if endSub == "NULL":
        string_sp +=  endSub +","
    else:
        string_sp += "'" + endSub+"',"
    
    distance = generate_distance()
    if distance == "NULL":
        string_sp +=  str(distance) +","
    else:
        string_sp += str(distance) +","
    
    qualifications = generate_qualifications()
    if qualifications == "NULL":
        string_sp +=  qualifications +","
    else:
        string_sp += "'" + qualifications+"',"
    
    if endSub == "NULL":
        string_sp += "1);\n"
    else:
        string_sp += str(randint(2,4)) + ");\n"
    file.write(string_sp)
```

```python
from random import randint

def create_users(number_of_users,file):
    i = 1
    
    #1 admin | 2 consumer | 3 sp | 4 company 
    
    while i < number_of_users+1:
        num = randint(0,100)
    
        if num > 90:
            create_company(i,file)
        elif num < 60:
            create_consumer(i,file)
        else:
            create_sp(i,file)
        i+=1
```
