```python
import names

def generate_name(sex):
    if sex == 'M':
        return names.get_full_name(gender='male')
    elif sex == 'F':
        return names.get_full_name(gender='female')
    else:
        return names.get_full_name()
```

```python
def generate_password(length):

    chars = "ABCDEFGHJKLMNPQRSTUVWXYZ0123456789"
    from os import urandom

    return "".join(chars[c % len(chars)] for c in urandom(length))
```

```python
def has_day(dias_escolhidos,dia):
    
    for i in dias_escolhidos:
        if i == dia:
            return "TRUE"
    
    return "FALSE"
```

```python
from random import randint

def create_sp(i,file):
    string_user = "INSERT INTO User(idUser,name,password,email,phoneNumber,sex,type,createdAt,lastActivity,active,idLocation)VALUES("
    string_user += str(i)+","
    
    sex = generate_sex()
    
    string_user += "'" + generate_name(sex) +"',"
    num_pw = randint(5,20)
    string_user += "'" + generate_password(num_pw) +"',"
    string_user += "'" + generate_email() +"',"
    
    phoneNumber = generate_phoneNumber()
    if phoneNumber == "NULL":
        string_user +=  phoneNumber +","
    else:
        string_user += "'" + phoneNumber+"',"
        
    string_user += "'" + sex+"',"
    string_user += str(3) +","
    string_date = generate_date().split("|")
    
    string_user += "'" + string_date[0]+"',"
    string_user += "'" + string_date[1]+"',"
    string_user +=  str(generate_active())+","
    string_user +=  str(generate_location())+");\n"
    
    file.write(string_user)

    
    string_sp = "INSERT INTO ServiceProvider(idSP,description,dateOfBirth,endSub,endSubVip,distance,qualifications,averageRating,workSchedule,occupiedSchedule,idSubscription)VALUES("
    string_sp += str(i)+","
    string_sp += "'" + generate_description() +"',"
    
    dateOfBirth = generate_dateOfBirth()
    if dateOfBirth == "NULL":
        string_sp +=  dateOfBirth +","
    else:
        string_sp += "'" + dateOfBirth+"',"
    
    endSub = generate_endSub()
    if endSub == "NULL":
        string_sp +=  endSub +","
        string_sp +=  endSub +","
    else:
        string_sp += "'" + endSub+"',"
        vip = randint(0,2)
        if vip == 0:
            string_sp += "'" + generate_endSubVip()+"',"
        elif vip == 1: 
            string_sp += "'" + endSub+"',"
        else:
            string_sp += "NULL,"
    
    distance = generate_distance()
    if distance == "NULL":
        string_sp +=  str(distance) +","
    else:
        string_sp += str(distance) +","
    
    qualifications = generate_description()
    if qualifications == "NULL":
        string_sp +=  qualifications +","
    else:
        string_sp += "'" + qualifications+"',"
    
    string_sp += "0,"
    
    auxiliar = create_work_schedule(i)
    
    string_sp += auxiliar+ ","
        
    if endSub == "NULL":
        string_sp += "1);\n"
    else:
        string_sp += str(randint(2,7)) + ");\n"
    file.write(string_sp)
```
