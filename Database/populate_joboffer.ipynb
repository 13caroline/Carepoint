```python
from random import randint

def generate_postDate():
    string = ""
    
    # Data de criacao
    ano = randint(2022,2023)
    mes = randint(1,12)
   
    if mes == 1 or mes == 3 or mes == 5 or mes == 7 or mes == 8 or mes == 10 or mes == 12:
        dia = randint(1,31)
    elif mes == 2:
        #fuck you 29
        dia = randint(1,28)
    else:
        dia = randint(1,30)

    hora = randint(0,23)
    minuto = randint(0,59)
    segundo = randint(0,59)
    
    string+=str(ano)+"-"
    if mes < 10: 
        string+= "0" + str(mes)+"-"
    else:
        string+= str(mes)+"-"
    
    if dia < 10: 
        string+= "0" + str(dia)+" "
    else:
        string+= str(dia)+" "
    
    if hora < 10: 
        string+= "0" + str(hora)+":"
    else:
        string+= str(hora)+":"
    
    if minuto < 10: 
        string+= "0" + str(minuto)+":"
    else:
        string+= str(minuto)+":"
    
    if segundo < 10: 
        string+= "0" + str(segundo)
    else:
        string+= str(segundo)
    
    return string
```

```python
from random import randint

def generate_price():
    
    num = randint(0,5)
    
    if num < 2:
        return "NULL"
    
    num = randint(0,99)
    
    if num < 10:
        return str(randint(5,20)) + ".0" + str(num)
    else:
        return str(randint(5,20)) + "." + str(num)
```

```python
from random import randint

def generate_aux(length):

    chars = "abcdefghijlmnopqrstuvxz "
    from os import urandom

    return "".join(chars[c % len(chars)] for c in urandom(length))

def generate_description():

    num = randint(1,1600)

    return generate_aux(num)
```

```python
def get_consumers_id():
    file_user = open("user.sql", "r")

    consumer = []
    i = 0
    for row in file_user:
   
        if i == 0:
            i+=1
        else:
            x = row.split(",")
            if x[2] == "password" and x[16] == '2':
                y = x[10].split('(')
                consumer.append(y[1])

    for i in consumer:
        print(i)
        
    file_user.close()
    return consumer
```

```python
def get_consumers_locations():
    file_user = open("user.sql", "r")

    location = []
    i = 0
    for row in file_user:
   
        if i == 0:
            i+=1
        else:
            x = row.split(",")
            if x[2] == "password" and x[16] == '2':
                y = x[20].split(')')
                location.append(y[0])

    for i in location:
        print(i)
        
    file_user.close()
    return location
```

```python
from random import randint

def create_joboffer(i,file,consumers,locations):
    j = 1
    tamanho_consumers = len(consumers)
    
    while j < i+1:
        string = "INSERT INTO joboffer(idJobOffer,description,beginDate,postDate,price,done,endDate,idUser,idCategory,idLocation)VALUES(" + str(j) + ",'"
        string += generate_description() + "','"
        string += generate_beginDate() + "','"
        string += generate_postDate() + "',"
        
        price = generate_price()
        if price == "NULL":
            string += price + ","
        else:
            string += str(price) + ","
            
        done = randint(0,1)
        string += str(done) + ","
        
        check = randint(0,10)
        if check < 3:
            string += "NULL" + ","
        else:
            string += "'" + generate_endDate() + "',"
            
        id_user = randint(0,tamanho_consumers - 1)
        string += str(consumers[id_user]) + ","
        string += str(randint(1,6)) + ","
       
        string += str(locations[id_user]) + ");\n"
        
        file.write(string)
        j+=1
    
    file.close()
```

```python
file = open("joboffer.sql", "a")
file.write("USE PI;\n");

#1 admin | 2 consumer | 3 sp | 4 company 
consumers = get_consumers_id()
locations = get_consumers_locations()

create_joboffer(100,file,consumers,locations)

file.close()
```
